## 1. Imports

```python
import numpy as np
import cv2
import os
import pprint
from tqdm import tqdm

import display_util

import matplotlib.pyplot as plt
import matplotlib.image as mpimg
%matplotlib inline
```

## 2. Primary settings

```python
SOURCE_DIR = "./../data/"
# IMAGE_NAME = "dwieFale.bmp"
IMAGE_NAME = "trojkat.bmp"
```

## 3. Loading images

```python
# building path to file
image_path = os.path.join(SOURCE_DIR, IMAGE_NAME)
# loading files
image = cv2.imread(image_path, 0)
```

```python
display_util.image(image, IMAGE_NAME, size=(7,7))
```

```python
f = np.fft.fft2(image)
spectrum = np.fft.fftshift(f)
amplitude = np.abs(spectrum)
magnitude = np.log(amplitude + 1)
phase = np.angle(spectrum)
```

```python
display_util.image(magnitude, IMAGE_NAME, size=(7,7))
```

```python
display_util.image(phase, IMAGE_NAME, size=(7,7))
```

```python
f_ishift = np.fft.ifftshift(spectrum)
img_back = np.fft.ifft2(f_ishift)
img_back = np.abs(img_back)
```

```python
display_util.image(img_back, IMAGE_NAME, size=(7,7))
```
