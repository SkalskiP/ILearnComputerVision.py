## Imports

```python
import numpy as np
import cv2
import os

import matplotlib.pyplot as plt
%matplotlib inline
```

## Settings

```python
source_dir = "./../data/"
file_names = [
    'lena1.bmp',
    'lena2.bmp',
    'lena3.bmp',
    'lena4.bmp'
]
```

## Loading images

```python
# building paths to files
file_paths = [os.path.join(source_dir, file) for file in file_names]
# loading files
files = [cv2.imread(path, 0) for path in file_paths]
```

## Display histograms

```python
f, axarr = plt.subplots(4, 2, figsize=(14, 20))
for index, file in enumerate(files):
    file_name = file_names[index]
    axarr[index, 0].imshow(file,'gray')
    axarr[index, 0].set_title(file_name)
    axarr[index, 1].hist(file.ravel(), 256, [0,256])
    axarr[index, 1].set_title("{} histogram".format(file_name))
```
