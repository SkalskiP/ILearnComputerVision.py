## 1. Imports

```python
import numpy as np
import cv2
import os
import pprint
from tqdm import tqdm

import display_util

import matplotlib.pyplot as plt
%matplotlib inline
```

## 2. Primary settings

```python
SOURCE_DIR = "./../data/"
IMAGE_NAMES = [
    "ferrari.bmp",
    "rice.png"
]
```

## 3. Loading images

```python
# building paths to images
images_paths = [os.path.join(SOURCE_DIR, file) for file in IMAGE_NAMES]
# loading files
images = [cv2.imread(path, 0) for path in images_paths]
```

### 3.1 Data overview

```python
display_util.images_with_histograms(files_list=images, names_list=IMAGE_NAMES)
```

## 4. Erosion and dilatation

### 4.1. Local settings

```python
KERNEL_SHAPE = (3, 3)
IMAGE = np.copy(images[0])
```

### 4.2. Implementation

```python
kernel = cv2.getStructuringElement(cv2.MORPH_RECT, KERNEL_SHAPE)
image_eroded = cv2.erode(IMAGE, kernel, iterations = 1)
image_dilated = cv2.dilate(IMAGE, kernel, iterations = 1)
```

```python
images_to_display = [
    IMAGE,
    image_eroded,
    image_dilated
]
images_names = [
    "original image",
    "eroded image",
    "dilated image"
]
display_util.images_with_histograms(files_list=images_to_display, names_list=images_names)
```

## 5. Opening and closing

### 5.1. Local settings

```python
KERNEL_SHAPE = (3, 3)
IMAGE = np.copy(images[0])
```

### 5.2. Implementation

```python
kernel = cv2.getStructuringElement(cv2.MORPH_RECT, KERNEL_SHAPE)
image_opening = cv2.morphologyEx(IMAGE, cv2.MORPH_OPEN, kernel)
image_closing = cv2.morphologyEx(IMAGE, cv2.MORPH_CLOSE, kernel)
```

```python
images_to_display = [
    IMAGE,
    image_opening,
    image_closing
]
images_names = [
    "original image",
    "opaned image",
    "closed image"
]
display_util.images_with_histograms(files_list=images_to_display, names_list=images_names)
```
