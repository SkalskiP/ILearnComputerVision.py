## Imports

```python
import numpy as np
import cv2
import os

import display_util

import matplotlib.pyplot as plt
%matplotlib inline
```

## Settings

```python
SOURCE_DIR = "./../data/"
IMAGE_NAMES = [
    "lenaSzum.bmp",
    "lena.bmp"
]
```

## Loading images

```python
# building paths to files
file_paths = [os.path.join(SOURCE_DIR, file) for file in IMAGE_NAMES]
# loading files
files = [cv2.imread(path, 0) for path in file_paths]
```

## Display lenaSzum.bmp

```python
IMAGE = files[0]
```

```python
plt.style.use('dark_background')
fig = plt.figure(figsize=(16,8))
fig.suptitle('Basic histogram', fontsize = 25)
ax1 = fig.add_subplot(121)
ax1.imshow(IMAGE, 'gray')
ax1.set_title("image")

ax2 = fig.add_subplot(122)
ax2.hist(IMAGE.ravel(), 256, [0,256], color = '#9E7CC1')
ax2.set_title("histogram")

plt.show()
```

## Median Blurring

### Settings

```python
IMAGE = files[0]
KERNEL_SIZE = 3
```

### Implementation

```python
output_image = cv2.medianBlur(IMAGE, KERNEL_SIZE)
```

```python
plt.style.use('dark_background')
fig = plt.figure(figsize=(16,8))
fig.suptitle('Median Blurring', fontsize = 25)
ax1 = fig.add_subplot(121)
ax1.imshow(output_image, 'gray')
ax1.set_title("image")

ax2 = fig.add_subplot(122)
ax2.hist(output_image.ravel(), 256, [0,256], color = '#9E7CC1')
ax2.set_title("histogram")

plt.show()
```
